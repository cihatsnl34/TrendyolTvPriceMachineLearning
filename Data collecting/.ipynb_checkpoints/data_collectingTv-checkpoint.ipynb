```python
#Gerekli kütüphaneler
import pandas as pd 
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
#Gerekli listeler
url_list = []
prices_list = []
propTitles = []
#Özelliklerin çekilmesi
for i in range(1,2): #2 yerine sayfa sayısı gelmeli
    url = "https://www.trendyol.com/sr?wc=104156&os=1&sk=1&pi=" + str(i) #Urlin for döngsü ile güncellenmesi
    r = requests.get(url) #Hazırlanan url'e istek yapılması
    source = BeautifulSoup(r.content,"lxml") #İstek yapılan sayfanın içeriğinin çekilmesi
    
    
    urls = source.find_all("div",attrs={"class":"p-card-chldrn-cntnr"}) #Tüm verilerin bulunması
    for url in urls:
        url_tv = "https://www.trendyol.com"+url.a.get("href") #Verilerin linklerinin bulunması
        url_list.append(url_tv) #Linklerin kaydedilmesi
        print(url_tv)
        
        r_tv = requests.get(url_tv) #Bulunan linke tekrar istek atılması
        source_phone = BeautifulSoup(r_tv.content,"lxml") #Veri içeriğinin çekilmesi
        
        properties = source_phone.find_all("li",attrs={"class":"detail-attr-item"}) #Bütün özelliklerin bulunması
        
        for prop in properties:
            prop_title = prop.find_all("span")[0].text
            propTitles.append(prop_title)
    
        
    prices = source.find_all("div",attrs={"class":"prc-box-sllng"}) #Bütün fiyat özelliklerinin bulunması
    for price in prices:
        prices_list.append(price.text) #Bulunan özellikler arasında gezinme ve listeye yazma
        print(price.text)
    
    
        
print(str(len(url_list))+" Adet link bulundu.")
print(str(len(prices_list))+" Adet fiyat bulundu.")
print(str(len(propTitles))+" Adet özellik başlığı bulundu.")
```
